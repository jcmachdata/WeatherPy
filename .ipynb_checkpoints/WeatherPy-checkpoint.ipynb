### Analysis
* As we can see from the Temperature v. Latitude plot, the temperature drops considerably the further north from the equator that a city is located. However, temperatures were consistently warm (about 60-90 deg F) in the southern hemisphere and equatorial region. This is most likely due to the current season. 
* There was no significant difference in cloud cover versus latitude, however there is an interesting trend involving the tendency of clouds to be present in specific percentages. They almost appear quanitized, like the energy levels of an atom. 
* Going from the southern hemisphere to the equator, the average wind speed drops precipitously. It then picks up again in the northern hemisphere and actually increases slightly.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#preview cities
cities
```

```python
#convert city data to dataframe
weather_df = pd.DataFrame(cities)
```

```python
weather_df.head()
```

```python
weather_df.columns=["City"]
```

```python
weather_df.head()
```

```python
#add needed columns to dataframe
weather_df["Country"]=""
weather_df["Lat"]=""
weather_df["Lng"]=""
weather_df["Date"]=""
weather_df["Cloudiness"]=""
weather_df["Humidity"]=""
weather_df["Max Temp"]=""
weather_df["Wind Speed"]=""
```

```python
weather_df.head()
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="

#loop through weather data, grab needed field values and add to dataframe
for index, row in weather_df.iterrows():
    city_name = row["City"]
    print(f"Retrieving Results for Index {index}: {city_name}.")
    response = requests.get(query_url + city_name).json()
      
    try:
    
        weather_df.set_value(index, "Country", response['sys']['country'])
        weather_df.set_value(index, "Lat", response['coord']['lat'])
        weather_df.set_value(index, "Lng", response['coord']['lon'])
        weather_df.set_value(index, "Date", response['dt'])
        weather_df.set_value(index, "Cloudiness", response['clouds']['all'])
        weather_df.set_value(index, "Humidity", response['main']['humidity'])
        weather_df.set_value(index, "Max Temp", response['main']['temp_max'])
        weather_df.set_value(index, "Wind Speed", response['wind']['speed'])
    
    except Exception as e:
        print(e)
        print("Missing data... skipping!")

```

```python
weather_df.head()
```

```python
#replace space filled values with NaN
weather_df = weather_df.replace("", np.nan, regex=True)
```

```python
#drop NaN rows
weather_df = weather_df.dropna()
```

```python
pprint(response)
```

```python
#Latitude v Temperature plot
plt.scatter(weather_df["Lat"], weather_df["Max Temp"], c = weather_df["Max Temp"], edgecolor="red", linewidths=1, marker="o", 
            alpha=0.7, label="City")

plt.style.use('seaborn')
plt.title(f"Max Temperature (F) vs. Latitude {datetime.now().strftime('%m/%d/%Y')}")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-80, 100])
plt.ylim([-40, 100])

# Save the figure
#plt.savefig("Latitude_v_Latitude.png")

# Show plot
plt.show()
```

```python
#Humidity v Latitude plot
plt.scatter(weather_df["Lat"], weather_df["Humidity"], c = weather_df["Humidity"], edgecolor="orange", linewidths=1, marker="o", 
            alpha=0.9, label="City")

#plt.style.use('seaborn')
plt.title(f"Humidity (%) vs. Latitude {datetime.now().strftime('%m/%d/%Y')}")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-80, 100])
plt.ylim([-5, 105])

# Save the figure
plt.savefig("Humidity_v_Latitude.png")

# Show plot
plt.show()
```

```python
#Cloudiness v Temperature plot
plt.scatter(weather_df["Lat"], weather_df["Cloudiness"], c = weather_df["Cloudiness"], edgecolor="blue", linewidths=1, marker="o", 
            alpha=0.7, label="City")

#plt.style.use('seaborn')
plt.title(f"Cloudiness (%) vs. Latitude {datetime.now().strftime('%m/%d/%Y')}")
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-80, 100])
plt.ylim([-5, 110])

# Save the figure
plt.savefig("Cloudiness_v_Latitude.png")

# Show plot
plt.show()
```

```python
#Wind Speed v Temperature plot
plt.scatter(weather_df["Lat"], weather_df["Wind Speed"], c = weather_df["Wind Speed"], edgecolor="green", linewidths=1, marker="o", 
            alpha=0.7, label="City")

#plt.style.use('seaborn')
plt.title(f"Wind Speed (mph) vs. Latitude {datetime.now().strftime('%m/%d/%Y')}")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-60, 90])
plt.ylim([-1, 40])

# Save the figure
plt.savefig("WindSpeed_v_Latitude.png")

# Show plot
plt.show()
```
